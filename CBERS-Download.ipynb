```python
import pandas as pd
from datetime import date
from cbers4asat import Cbers4aAPI

# Função para realizar a consulta e converter para GeoDataFrame
def consulta_cbers(api, path_row, data_inicial, data_final):
    produtos = api.query(location=path_row,
                         initial_date=data_inicial,
                         end_date=data_final,
                         cloud=100,
                         limit=100,
                         collections=['CBERS4A_WPM_L4_DN'])
    return api.to_geodataframe(produtos, 'EPSG:4674')

api = Cbers4aAPI('hugonbgg@email.com')

# Lista de path rows para consultar
path_rows = [(202, 142), (202, 143), (203, 142), (203, 143)]

data_inicial = date(2023, 1, 1)
data_final = date(2024, 12, 1)

# Lista para armazenar os GeoDataFrames de cada consulta
gdfs = []

# Realiza a consulta para cada path row e armazena o GeoDataFrame na lista
for path_row in path_rows:
    gdf = consulta_cbers(api, path_row, data_inicial, data_final)
    gdfs.append(gdf)

# Concatena todos os GeoDataFrames em um único GeoDataFrame
gdf_final = pd.concat(gdfs, ignore_index=True)

# Converte a coluna 'datetime' para o tipo datetime
gdf_final['datetime'] = pd.to_datetime(gdf_final['datetime'])

# Adiciona colunas para ano e mês
gdf_final['year'] = gdf_final['datetime'].dt.year
gdf_final['month'] = gdf_final['datetime'].dt.month

# Função para encontrar a melhor imagem para cada grupo de 4 path rows por ano e mês
def find_best_group(df):
    best_group = pd.DataFrame()
    for (year, month), group in df.groupby(['year', 'month']):
        group = group.sort_values('cloud_cover')
        # Filtra os primeiros 4 com menor cobertura de nuvens
        group = group.groupby(['path', 'row']).first().reset_index()
        if len(group) == 4:
            best_group = pd.concat([best_group, group])
    return best_group

best_images = find_best_group(gdf_final)

# Exibe as melhores imagens agrupadas por ano e mês
best_images
```
